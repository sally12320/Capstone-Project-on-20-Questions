```python
import pandas as pd
import numpy as np
import copy
import random
df = pd.read_csv('change_binary_fin.csv')
df
```

```python
def func_entropy(ser):
    p = ser.value_counts(normalize=True)
    return -np.sum(p * np.log2(p))
```

```python
def ask_question_and_narrow_down(df, previous_responses={}):
    # If only 1 name left, find answer
    if len(df) == 1:
        # f1 = True
        print(f"The character you are thinking of is: {df['Name'].values[0]}")
        return df['Name'].values[0]

    # calculate CMI for remaining columns
    cmis =  df.iloc[:,1:].apply(func_entropy).to_dict()
    
    # check if all CMIs are equal, if so just return 
    if len(set(cmis.values())) == 1:
        # fcmis = True
        print("All characters remaining have the same information gain for remaining questions.")
        print("The characters you pick are among:")
        names = df['Name'].to_list()
        names_string = ', '.join(names)
        print(names_string)
        return names_string
    
    # find best col to ask according to enropy calc
    best_question = df.iloc[:,1:].apply(func_entropy).sort_values(ascending=False).head(1).index[0]
    
    # take in user input
    response = input(f"Does the character {best_question}? (yes/no): ").strip().lower()
    
    # update 
    if response == 'yes':
        previous_responses[best_question] = 1
        df = df[df[best_question] == 1]
    elif response == 'no':
        previous_responses[best_question] = 0
        df = df[df[best_question] == 0]
    else:
        print("Please answer with 'yes' or 'no'.")
        return ask_question_and_narrow_down(df, previous_responses={})

    # print names remaining after each question being asked 
    print(f"Characters remaining: {len(df)}")
    
    # when only 10 or less names left, print names
    if len(df) <= 10:
        print("Remaining characters:")
        print(df['Name'].to_list())

    # recursive call
    ask_question_and_narrow_down(df, previous_responses={})
```
